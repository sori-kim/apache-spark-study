```python
from pyspark import SparkConf, SparkContext
```

```python
import re
```

```python
def normalizeWords(text):
    return re.compile(r'\W+', re.UNICODE).split(text.lower())
```

```python
conf = SparkConf().setMaster('local').setAppName('WordsCount') 
```

```python
sc = SparkContext.getOrCreate();
```

```python
sc
```

```python
data = sc.textFile('File:///Users/rosie/Desktop/sparkCourse/book.txt')
```

```python
words = data.flatMap(normalizeWords)
```

```python
wordCounts = words.map(lambda x: (x, 1)).reduceByKey(lambda x,y: x + y)
```

```python
wordCounts.collect()
```

```python
wordCountsSorted = wordCounts.map(lambda xy: (xy[1],xy[0])).sortByKey()
```

```python
results = wordCountsSorted.collect()
```

```python
results
```

```python
for result in results:
    count = str(result[0])
    word = result[1].encode('ascii', 'ignore')
    if(word):
        print (word , ":\t\t" + count)
```
