```python
from pyspark import SparkConf, SparkContext
```

```python
conf = SparkConf().setMaster('local').setAppName('CustomerSpent')
```

```python
sc = SparkContext.getOrCreate()
```

```python
data = sc.textFile('File:///Users/rosie/Desktop/sparkCourse/customer-orders.csv')
```

```python
data.collect()
```

```python
def parseLine(data):
    parsed = data.split(',')
    customerId = int(parsed[0])
    spent = float(parsed[2])
    return (customerId, spent)
```

```python
parsedData = data.map(parseLine).reduceByKey(lambda x,y: x + y)
```

```python
sortedData = parsedData.map(lambda xy: (xy[1],xy[0])).sortByKey()
```

```python
results = sortedData.collect()
```

```python
for result in results:
    print (result)
```
